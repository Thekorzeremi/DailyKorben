## Project
### Selenium, dynamic approach
First of all, we need to import our libraries :

````python
import json # save scrapped datas in a file
from selenium import webdriver # driver to control the selenium browser instance
from selenium.webdriver.common.by import By # Needed to access to specific website content to scrap
from selenium.webdriver.chrome.options import Options # Start browser instance with specific options
````

Now let's just write some code to scrapped our website !  

````python
options = Options() # creating our options object
options.add_argument('--headless=new') # add headless argument to our options object to start browser without visible UI for user
driver = webdriver.Chrome(options=options) # creating the Chrome driver instance for selenium

url = "https://korben.info/" # the url to scrap
driver.get(url) # telling the driver to go on the URL

elements = driver.find_elements(By.CLASS_NAME, 'article-card-content') # catching website multiple sections thanks to their classname (article-card-content)

scrapped_data = {"data": []} # creating a variable to store retrieve data
count = 1 # creating a count variable to check the number of articles to scrap

# We just want to scrap the latest three articles from the website because Korben daily post three articles

# creating a loop on catching sections to add the latest three articles to scrapped_data
for e in elements:
    # only keeping title and content
    if count < 4:
        raw = e.text.strip()
        lines = raw.split("\n")

        if len(lines) >= 5:
            title = lines[0]
            content = "\n".join(lines[4:])
        else:
            title = lines[0]
            content = ""

        scrapped_data["data"].append({
            "title": title,
            "content": content
        })

        count += 1

driver.quit() # Quit driver instance to stop py script
````

Now we have scrapped three articles ! Let's check with print command :

````python
print(scrapped_data)
````
